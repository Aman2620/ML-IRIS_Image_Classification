```python
training_data = training_data.rename(columns={0:'sepal length',1:'sepal width',2:'petal length',3:'petal length'})
```

```python
training_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = training_data.iloc[:,:4]
target_features = training_data.iloc[:,4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,target_features,test_size=0.2,stratify = target_features)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, prediction))
```
